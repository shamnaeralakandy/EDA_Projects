### mental_health_by_gender

```python
mental_health_by_gender = df.groupby('Gender')[['Anxiety_score', 'Depression_score', 'Stress_level']].mean()
mental_health_by_gender
```

```python
mental_health_by_gender_sorted = mental_health_by_gender.sort_values(by=['Anxiety_score', 'Depression_score'])
mental_health_by_gender_sorted
```

```python
plt.figure(figsize=(5,4))
mental_health_by_gender.plot(kind="bar")
plt.title("mental health distribution by gender")
plt.xlabel("Gender")
plt.ylabel("mental health")
plt.tight_layout()
plt.show
```

- Non-Binary individuals report the highest average scores in Anxiety,Depression,Stress

- Males and Females show similar levels, with:Males having slightly higher depression, females slightly higher anxiety.stress levels are almost equal.

- The "Other" gender group shows the lowest average anxiety and depression scores, but slightly higher stress than males/females.

### Relationship between Anxiety, Depression, and Stress Scores

```python
corr_mental_health_score=df[['Anxiety_score','Depression_score', 'Stress_level']].corr()
corr_mental_health_score
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(corr_mental_health_score,annot=True,cmap="crest")
plt.title("Correlation between Anxiety, Depression, and Stress Scores")
plt.show()
```

- There is no significant linear correlation between anxiety, depression, and stress scores in this dataset.
- The mental health factors appear to vary independently from one another here

### Correlation of Self-Esteem with Mental Health Scores

```python
mental_health_cols = ['Anxiety_score', 'Depression_score', 'Stress_level']
correlations = df[mental_health_cols + ['Self_esteem_score']].corr()

sns.heatmap(correlations[['Self_esteem_score']], annot=True, cmap='YlGnBu')
plt.title("Correlation of Self-Esteem with Mental Health Scores")
plt.show()
```

- As self-esteem increases, there is a slight decrease in anxiety and depression levels.
- However, the relationship is very weak, suggesting other factors may have a stronger influence.
- The correlation between self-esteem and stress is close to zero (+0.012), indicating no meaningful linear relationship

### Correlation between Loneliness and Life Satisfaction

### Sleep Hours vs Mental Health

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(data=df, x="Sleep_hours", y="Depression_score")
sns.regplot(data=df, x="Sleep_hours", y="Depression_score", scatter=False, color='red')
plt.title("Depression vs Sleep Duration")

plt.subplot(1, 3, 2)
sns.scatterplot(data=df, x="Sleep_hours", y="Anxiety_score")
sns.regplot(data=df, x="Sleep_hours", y="Anxiety_score", scatter=False, color='red')
plt.title("anxiety vs Sleep Duration")

plt.subplot(1, 3, 3)
sns.scatterplot(data=df, x="Sleep_hours", y="Stress_level")
sns.regplot(data=df, x="Sleep_hours", y="Stress_level", scatter=False, color='red')
plt.title("stress level vs Sleep Duration")

plt.show()
```

There is no strong linear relationship between sleep duration and levels of depression, anxiety, or stress. However, there is a slightly decreasing trend in stress levels with increasing sleep hours, suggesting that getting more sleep may have a mild stress-reducing effect, although the impact appears limited.

### Mental Health by Medication Use

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
sns.violinplot(data=df, x="Medication_use", y="Stress_level", inner="box", palette="Pastel1")
plt.title("Stress by Medication_use")

plt.subplot(1, 3, 2)
sns.violinplot(data=df, x="Medication_use", y="Anxiety_score", inner="box", palette="deep")
plt.title("Anxiety by Medication_use")

plt.subplot(1, 3, 3)
sns.violinplot(data=df, x="Medication_use", y="Depression_score", inner="box", palette="Set2")
plt.title("Depression by Medication_use")

plt.tight_layout()
plt.show()
```

There is no substantial difference in stress, anxiety, or depression scores across medication usage levels (None, Occasional, Regular). This suggests that medication use may not be significantly altering mental health outcomes, or that individuals may be using medication in response to symptoms rather than it reducing them significantly.

## conclusion

- Bachelor’s degree holders have the highest depression scores.
- Master’s degree holders show the highest anxiety levels.
- Individuals with a PhD report the lowest average anxiety and depression scores.
- Stress levels are fairly stable across all education levels, slightly higher among Master’s and “Other” degree holders.
- Anxiety and depression scores vary across age groups, peaking in the mid-30s and early 60s.
- Middle-aged individuals (30s–40s) show slightly higher mental health concerns.
- Stress levels remain consistent across age groups, with small variations.
- Students report the highest depression scores, while unemployed individuals show slightly higher anxiety and depression than others.
- Stress scores are similar across employment statuses.
- Anxiety, depression, and stress do not show strong linear correlations and seem to vary independently.
- Higher self-esteem is slightly associated with lower anxiety and depression, though the relationship is weak.
- Meditation and therapy users tend to have slightly better mental health scores.
- Increased sleep hours show a minor reduction in stress but little impact on anxiety or depression.
- Frequent substance users have slightly higher depression scores; anxiety and stress are not significantly affected.
- Medication use (none, occasional, regular) shows no major impact on mental health outcomes.
- Social support levels do not significantly influence anxiety, depression, or stress in this dataset.
- Work stress and financial stress scores do not show strong or consistent effects on mental health indicators.
- Overall, internal and lifestyle factors (like self-esteem, therapy, and sleep) show more consistent relationships with mental health outcomes - than external stressors (like work or financial pressures).
